```python
from preprocessor.missing_value_analysis import MissingValueAnalyzer
from data_injestion import train_df,test_df
from preprocessor.imputer import Preprocessor

sparse_columns = train_df.isnull().mean()[train_df.isnull().mean() > 0.8].index.to_list()
preprocessor = Preprocessor(cardinality_threshold=10, skew_threshold=0.5,sparse_columns=["Id"])
df_imputed = preprocessor.fit_transform(train_df)
# df_imputed.head()
```

```python
from feature_engineering.feature_extraction import NumericCorrelationSelector, CategoricalCorrelationSelector, CorrelationVisualizer
```

```python
from sklearn.pipeline import Pipeline
from feature_engineering.feature_extraction import NumericCorrelationSelector, CategoricalCorrelationSelector, CorrelationVisualizer

pipe = Pipeline([
    ("visualize", CorrelationVisualizer(target="SalePrice")),
    ("num_selector", NumericCorrelationSelector(target_col="SalePrice", corr_threshold=0.7)),
    ("cat_selector", CategoricalCorrelationSelector(target="SalePrice", eta_threshold=0.25, pvalue_threshold=0.05))
])

pipe.fit(df_imputed)               # computes correlations and selects features
target = getattr(pipe.named_steps["visualize"],"target","SalePrice")

pipe.named_steps["visualize"].plot()  # shows numeric + categorical plots

# Get selected numeric/categorical features
num_features = pipe.named_steps["num_selector"].selected_features_
cat_features = pipe.named_steps["cat_selector"].selected_features_
print(num_features)
print(cat_features)
print(target)
```

```python
df = df_imputed[num_features + cat_features + [target]]
```
